### We will define a function named bubble_sort(to_be_sorted) to perform the bubble sort. A function definition in python is simple. The syntax is given below:


```python
def bubble_sort(to_be_sorted):
    for i in range(len(to_be_sorted),1,-1):  # outer loop start from the size of the list and decrement down
        for j in range(0,i-1,1):
            if to_be_sorted[j] > to_be_sorted[j+1]:   # compare adjacent elements, is left greater then right
                to_be_sorted[j],to_be_sorted[j+1] = to_be_sorted[j+1],to_be_sorted[j] # swap if left > right
    return to_be_sorted


a = [9,5,6,3,1]
b = bubble_sort(a)
print(b)
```

prompt user for a year and check if it is a leap year or not.
    Use the following steps to determine whether a year is a leap year or not.

If a year is evenly divisible by 4 then go to next step. If it is not divisible by 4. It is not a leap year. 
If a year is divisible by 4, but not by 100. It is a leap year. If a year is divisible by both 4 and 100, go to next step.
If a year is divisible by 100, but not by 400. For example: 1900, then it is not a leap year. If a year is divisible by both, then it is a leap year. Some of the leap years are 1820,1940,1960,2280.


```python
flag = True
while flag:
    a = int(input("Enter the year: "))
    if a%4 == 0:
        if a%100 !=0:
            print(f"{a} is a leap year")
            break
        else:
            print(f"{a} is divisible by 4 and 100")
            if a%400 !=0:
                print("It is not a leap year( cannot divide by 400, 100 and 4 can)")
            else:
                print("it is a leap year as can be divisible by 4,100,400")
                break
    else:
        print(f"{a} is not a leap year as it is not divisible by 4")
```
