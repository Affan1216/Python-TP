### Aliasing may give you problem that is hard to detect in your code. Aliasing means two or more names refer to the same object. <font color=red> It is dangerous to have alias for mutable data type. </font>Let us see an example below.

```python
a=[1,2,3,4]
```

```python
c = a  # assign a to c
```

```python
c is a  # a and c both refer to the same list, c is just an alias of a
```

```python
c
```

```python
id(a)
```

```python
id(c)   # same id means as memory location means same object
```

```python
c.append(5)  # add value 5 to c
```

```python
c
```

```python
a   # what happens to a? It get changed also. a and c refer to the same object!
```

```python
d = a[:]  # this is one way to clone an object, d is a new object having the same values as a
```

```python
d
```

```python
id(d)  # d is located at a different memory location from a
```

```python
d.append(6) # add 6 to d, will 6 be in a also?
```

```python
d
```

```python
a    # a is not changed because a and d are different objects
```

```python
a=[1,2,3,4]
c = a  # assign a to c
print(a)
print(c)
c.append(5) # add 5 to the end of list c
print(c)
print(a)    # a also get added with 5
d=a[:]      # d is a new object having the same values as a
```

### <font color=red>*Combine what you have done and create a function named nric_checker. It will take a NRIC number starting with 'S' as the parameter. It will check if the check reference of the NRIC is correct. It returns True if it is valid and False if it is not valid

```python
def nric_checker(nric):
    weight = (2,7,6,5,4,3,2)
    reference = ('A','B','C','D','E','F','G','H','I','Z','J')
    ref = nric[-1]
    sum_total = 0
    nric_digit = nric[1:8]
    for i in range(len(nric_digit)):
        sum_total = sum_total + int(nric[i]*weight[i])
    ref_index = 11 - sum_total%11
    
    if ref == reference[ref_index-1]:
        return True
    return False

nric_number = input("What is your Nric number? ").strip()
print(nric_checker(nric_number))
```

```python
def is_sorted(a_list, direction='asc'):
    if direction not in ['asc','dsc']:
        return "Unknown sorting direction"
    if direction =='dsc':
        return False
    if a_list == sorted(a_list, reverse=False):
        return True
    return False

my_list = [1,4,6,2,9]
print(is_sorted(my_list,'asc'))
```
