### The follow code uses 2 stacks to implement word reversing. We push the individual letter of the word into my_stack. We then pop out the letters in my_stack and push them into my_stack_reverse. Then we print out my_stack_reverse.

```python
my_stack=[]          # use a list to implement a stack
my_stack_reverse=[]  # this list will build the reversed word
word="DSAG"          # define a word
for s in word:       # append the letter in the word to my_stack
    my_stack.append(s)
for i in range (len(my_stack)):
    my_stack_reverse.append(my_stack.pop()) #take out the letters from my_stack and push into my_stack_reverse
print(my_stack_reverse)
```

### <font color=red> Let us create a Stack class

```python
class Stack:
    def __init__(self):
        # Initialize an empty list to represent the stack
        self.items = []

    def is_empty(self):
        # Check if the stack is empty
        return len(self.items) == 0

    def push(self, item):
        # Add an item to the top of the stack
        self.items.append(item)

    def pop(self):
        # Remove and return the item from the top of the stack
        if not self.is_empty():
            return self.items.pop()
        else:
            raise IndexError("Pop from an empty stack")

    def peek(self):
        # Return the item from the top of the stack without removing it
        if not self.is_empty():
            return self.items[-1]
        else:
            raise IndexError("Peek into an empty stack")

    def size(self):
        # Return the number of items in the stack
        return len(self.items)

def postfix(expression):
    stack = Stack()
    operand = ('+-,*/')
    
    for a in expression:
        if a not in operand:
            stack.push(a)
        else:
            op1 = stack.pop()
            op2 = stack.pop()
            result = eval(op1+a+op2)
            
            stack.push(str(result))
    return result
a = '42+352-*+'
b = '953*+4-5+'
print(postfix(a))
print(postfix(b))
    
```

### Let us create a Stack object and use the stack

```python
#create an object of Stack (instantiate an object)
my_stack = Stack()

# Push items onto the stack
my_stack.push(1)
my_stack.push(2)
my_stack.push(3)

# Check the size of the stack
print("Stack size:", my_stack.size())

# Peek at the top item
print("Top item:", my_stack.peek())

# Pop items from the stack
print("Popped item:", my_stack.pop())
print("Popped item:", my_stack.pop())

# Check the size of the stack again
print("Stack size:", my_stack.size())
```
